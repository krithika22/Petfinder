```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
train =pd.read_csv("train.csv",na_values=['','nan','-','nil','null','?','_','.'])
train.head(5)
train.shape[0]
```

```python
plt.figure(figsize=(10, 6));
sns.set(style="darkgrid")
sns.set_palette('dark')
ax = sns.countplot(x="AdoptionSpeed", data=train)
ax.set_title('Frequency plot for Adoption Speed')
```

```python
plt.figure(figsize=(6, 6));
sns.set(style="darkgrid")
sns.set_palette('dark')
ax = sns.countplot(x="Type", data=train)
ax.set_title('Frequency plot for Type')
```

```python
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
plt.figure(figsize=(10, 10))
stopwords = set(STOPWORDS)
stopwords.update(["dog", "cat"])
text = ' '.join(train.loc[(train['AdoptionSpeed'] == 0) ^ (train['AdoptionSpeed'] == 1),'Description'].fillna('').values)
wordcloud = WordCloud(stopwords=stopwords,background_color='white',
                      width=1500, height=1000).generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud of description');
plt.axis("off");
```
